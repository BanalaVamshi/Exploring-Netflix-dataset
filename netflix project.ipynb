```python
netflix['date_added'] = pd.to_datetime(netflix['date_added'])
netflix['year_month'] = netflix['date_added'].dt.to_period('M')
monthly_releases = netflix.groupby(['year_month', 'type']).size().unstack().fillna(0)
```

```python
monthly_releases.plot(kind='line', figsize=(14, 7), marker='o')
plt.title('Monthly Releases of Movies and TV Shows on Netflix')
plt.xlabel('Month-Year')
plt.ylabel('Number of Releases')
plt.legend(title='Type')
plt.grid(True)
plt.show()
```
